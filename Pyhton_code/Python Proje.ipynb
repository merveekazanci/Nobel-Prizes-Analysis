```python
# Yılları filtreleme (1947-1991 arası)
filtered_data = nobel_data[(nobel_data['year'] >= 1947) & (nobel_data['year'] <= 1991)]

# Kategoriler için ayrı ayrı grafikler oluşturma
categories = filtered_data['category'].unique()

fig, axes = plt.subplots(len(categories), 1, figsize=(14, len(categories) * 6))

for i, category in enumerate(categories):
    category_data = filtered_data[filtered_data['category'] == category]
    country_counts = category_data['birth_country'].value_counts()
    
    sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10], palette='viridis', ax=axes[i])
    axes[i].set_title(f'{category} Kategorisinde 1947-1991 Arası Nobel Ödülü Kazanan Ülkeler')
    axes[i].set_xlabel('Ülkeler')
    axes[i].set_ylabel('Kazanan Sayısı')
    axes[i].grid(True)
    
plt.tight_layout()
plt.show()
```

```python
#Yorum: ABD her kategoride öne çıkmış olduğunu görmekteyiz. 
Burdan şu sonucu çıkarabiliriz, ABD her dalda kendini geliştiren uzman kişilerin bulunduğu bir ülkedir.
Genel olarak diğer ülkelerinde ortak olarak edebiyat dalında diğer kategorilere göre daha 
fazla ödül kazandığı görülmektedir.Siyası açıdan bakmak gerekirse bu dönem Soğuk Savaş 
Dönemi olarak geçmektedir. ABD II.Dünya Savaşı ve Soğuk Savaş Dönemi zamanında hep gelişime 
açık olmuş ve bu dönemlerde nobel ödüllerinde hep brinci sırada yer almıştır.
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
 
file_nobel= r"C:\Users\merve\Downloads\asset-v1_Kodlasam+MS01+2023_06+type@asset+block@nobel.csv"
nobel_data= pd.read_csv(file_nobel)
```

```python
data_200 =nobel_data[nobel_data['year'] > 2000] 
data_200['age_at_award'] = data_200['year'] - pd.to_datetime(data_200['birth_date']).dt.year 
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine'] 
 
for category in categories: 
    category_list = data_200[data_200['category'] == category] 
 
    plt.figure(figsize=(12, 6)) 
 
    plt.subplot(1, 2, 1) 
    sns.countplot(y='birth_country', data=category_list, order=category_list['birth_country'].value_counts().iloc[:10].index) 
    plt.title(f'{category} Kategorisi - İlk 10 Ülke') 
    plt.xlabel('Kişi Sayısı') 
    plt.ylabel('Ülke') 

    plt.subplot(1, 2, 2) 
    sns.histplot(category_list['age_at_award'], bins=10, kde=True) 
    plt.title(f'{category} Kategorisi - Yaş Dağılımı') 
    plt.xlabel('Yaş') 
    plt.ylabel('Yoğunluk') 
 
    plt.tight_layout() 
    plt.show()
```

```python
#Yorum: Genel olarak yaş tablosuna bakıldığında her alanda 60 ve üzeri yaştaki kişilerin çoğunlukta ödül 
aldığı söylenebilir. Bu çıkarımdan yola çıkarak alanında deneyimli kişilerin uzun soluklu çalışmalar 
yapan kişilerin ödüle layık görüldğünü söylenebilir. Ülke bazında bakıldığında genel olarak ABD nobel 
ödüllerinde birici sırada yer almaktadır.Bu sonuca göre ABD için gelişime açık olduğu ve yatırımlarının 
karşılığını aldığını söyleyebiliriz. 
```
